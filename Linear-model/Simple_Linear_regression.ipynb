```python
import matplotlib.pyplot as plt
import numpy as np

rng = np.random.RandomState(42)
x = 10 * rng.rand(40)
y = 2 * x - 1 + rng.randn(40)

plt.scatter(x, y)
```

```python
# i) choose a class of model
from sklearn.linear_model import LinearRegression

"""
ii) choose hyper parameters
1.normalization, decide whether normalize input or input
2.Fit intercept, decide whether to calculate the intercept for the model
3. regularization, choose the degree of regularization to use in the model, regularization prevents overfitting
4. Feature Pre-processing,  Decide whether to pre-process the features to add model flexibility.
5. Solver
"""

model = LinearRegression(fit_intercept= True)
LinearRegression()
```

```python
"""
iii) Arrange data into features matrix and target vector

2D feature matrix and 1D target array

reshaping data x into two-dimensional array: make it a matrix of size[n_samplpes, n_features]

"""
X = x[:, np.newaxis]
X.shape
```

```python
"""
ix) fit the model to your data

"""
model.fit(X, y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
plt.title("X vs Y")
plt.scatter(x, y)
plt.plot(x, (model.coef_*x) + (model.intercept_), color = 'green')

plt.xlabel("X")
plt.ylabel("Y")
```

```python
"""
v) predict labels for unknown data
 use predict()
"""

xfit = np.linspace(-1, 11, 40)

# change the x values into a [n_samples, n_features] features matrix format
Xfit = xfit[:, np.newaxis]

yfit = model.predict(Xfit)
```

```python
"""
vi) check the result of model fitting to know whether the model is satisfactory

Model Performance(Optimization)
goodness of fit determies how the line of regression fits the set of observations.
use R-squared method(R2) - measures how well unseen samples are likely to be predicted by the model.
"""

r_sq = model.score(X, y)
print("cooeffiecient of determination: ", r_sq)
```
